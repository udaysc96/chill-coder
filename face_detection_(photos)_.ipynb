THE LIBRARIES USED IN THE CODES.

```
import numpy as np                #CONVERTING THE PHOTOS INTO ARRAYS AND ARRAY WORKING 
import cv2                        #COMPUTER VISION LIBRARY USED FOR THE IMAGE READING, WRITING, PROCESSING ETC.
import matplotlib.pyplot as plt   #USED FOR THE IMAGE PLOTING
```

THE BELOW CODE IS FOR THE FACE DETECTION IN THE IMAGE USED. THE RESULT WILL BE THE 4 COORDINATES OF THE FACE IN THE IMAGE.  

```
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
img = cv2.imread('smw.JPG')                                                    
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces = face_cascade.detectMultiScale(gray, 1.3, 5)
print(faces)
```

THE BELOW CODE IS TO MAKE ROI OF FACES AND TO PUT RECTANGLE IN THE IMAGE AROUND THE FACE. THE RESULT WOULD SHOW THE ACTUAL IMAGE BUT WITH RECTANGLE AROUND THE DETECTED FACES.

```
for x,y,w,h in faces:                              # X,Y ARE THE STARTING COORDINATE FOR THE FACES AND W,H ARE THE WIDTH AND HEIGHT RESPECTIVELY.
  cv2.rectangle(img, (x,y), (x+w,y+h), (0,255,0))  # PUT RECTNAGLE AROUND COORDINATES, COLOR AND WIDTH OF RECTANGLE CAN BE CHANGED
  roigray = gray[y:y+h,x:x+w]                      # GRAYCALE (FACE AREA)     
  roicolor = img[y:y+h,x:x+w]                      # COLOR (FACE AREA) 
plt.grid(None)
plt.xticks([])
plt.yticks([])
imgplot = plt.imshow(img)                          # PLOT THE WHOLE IMAGE WITH THE RECTANGLE AROUND THE FACE
```
